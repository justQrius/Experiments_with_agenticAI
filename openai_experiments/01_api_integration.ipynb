# OpenAI Agents SDK: Basic Implementation

This notebook explores the experimental OpenAI Agents SDK, demonstrating how to define lightweight agents with specific instructions and trace their execution.

```python
# Import dependencies
from dotenv import load_dotenv
from agents import Agent, Runner, trace
```

```python
# Initialize Environment
load_dotenv(override=True)
```

```python
# Define Agent
agent = Agent(
    name="Jokester", 
    instructions="You are a witty comedian. Generate sophisticated, tech-focused humor.", 
    model="gpt-4o-mini"
)
```

```python
# Execute Agent with Tracing
with trace("Agent Execution Trace"):
    result = await Runner.run(agent, "Tell a joke about Autonomous AI Agents")
    print(f"Agent Output: {result.final_output}")
```

### Observability

Execution traces can be viewed in the [OpenAI Platform Traces](https://platform.openai.com/traces) dashboard.
